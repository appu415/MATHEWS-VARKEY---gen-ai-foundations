```python
import pandas as pd
import numpy  as np

from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df = pd.read_csv("Reviews.csv", nrows=50000)
```

```python
light_reviews = df[df['Text'].str.contains(r'\blight\b', case=False, na=False)]

print("Number of reviews containing 'light':", len(light_reviews))
light_reviews[['Text']].head(20)
```

```python
sentences = [
    "This olive oil is wonderful and very light on the stomach.",
    "The packaging was broken and let too much light inside the box.",
    "I prefer this brand because it is a light and healthy snack."
]
```

```python

vectorizer = CountVectorizer()

dtm = vectorizer.fit_transform(sentences)
```

```python
dense_matrix = dtm.toarray()

print(dense_matrix)
```

```python
vec_1 = dense_matrix[0]
vec_2 = dense_matrix[1]
vec_3 = dense_matrix[2]

print("vec_1 =", vec_1)
print("vec_2 =", vec_2)
print("vec_3 =", vec_3)
```

```python


similarity_1_2 = cosine_similarity(
    [vec_1],
    [vec_2]
)

print("Similarity between Sentence 1 and Sentence 2:")
print(similarity_1_2[0][0])
```

```python
# Dot Product Similarities
similarity_1_2 = np.dot(vec_1, vec_2)
similarity_1_3 = np.dot(vec_1, vec_3)

# Console Output
print("Similarity between Sentence 1 and Sentence 2 (Different Contexts):")
print(similarity_1_2)

print("\nSimilarity between Sentence 1 and Sentence 3 (Similar Contexts):")
print(similarity_1_3)
```
